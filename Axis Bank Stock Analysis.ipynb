#### Q4. How did the stock perform before and after significant events, such as market crashes?

```python
events = {
    '2008 Financial Crisis': pd.Timestamp('2008-09-15'),
    'COVID-19 Market Crash': pd.Timestamp('2020-03-16')
}

def performance(df,event_date,days_before=30,days_after=30):
    start_date=event_date-pd.Timedelta(days=days_before)
    end_date=event_date+pd.Timedelta(days=days_after)


    before_event=df[(df.index >= start_date) & (df.index < event_date)]
    after_event= df[(df.index > event_date) & (df.index <= end_date)]

    # calculate average

    before_avg=before_event[['Close','Low','High','VWAP']].mean()
    after_avg=after_event[['Close','Low','High','VWAP']].mean()

    return before_avg,after_avg

performance_results={}

for event_name, event_date in events.items() :
    before_avg,after_avg=performance(axis_stock_df,event_date)
    
    performance_results[event_name]={
        'Before Event': before_avg,
        'After Event':after_avg
    }


for events,results in performance_results.items():
    print(f"\nPerformance around {events}:")
    print("Average performance before the event:")
    print(results['Before Event'])
    print("\nAverage performance after the event:")
    print(results['After Event'])
```

` According to the analysis we can see that after 2008 market crash axis bank stocks doesn't affected too much. But after Covid 19 market crash in 2020 there is strong changes in stock price.`

#### Q5. How often did the stock close above or below its opening price?

```python
# Calculate the difference between Close and Open prices
axis_stock_df['Close_Open_Diff'] = axis_stock_df['Close'] - axis_stock_df['Open']

# Count the number of days where the stock closed above or below the opening price
above_open_count = (axis_stock_df['Close_Open_Diff'] > 0).sum()
below_open_count = (axis_stock_df['Close_Open_Diff'] < 0).sum()

# Total number of trading days
total_days = len(axis_stock_df)

print(f"Days closed above the opening price: {above_open_count}")
print(f"Days closed below the opening price: {below_open_count}")
print(f"Total trading days: {total_days}")

# Calculate percentages
percentage_above_open = (above_open_count / total_days) * 100
percentage_below_open = (below_open_count / total_days) * 100

print(f"Percentage of days closed above the opening price: {percentage_above_open:.2f}%")
print(f"Percentage of days closed below the opening price: {percentage_below_open:.2f}%")
```

#### Q6. What are the top 5 periods of highest volatility in the stock price?

```python
import pandas as pd

# Assuming axis_stock_df is already prepared

# Calculate daily volatility as High - Low
axis_stock_df['Daily_Volatility'] = axis_stock_df['High'] - axis_stock_df['Low']

# Top 5 periods with highest daily volatility
top_5_volatility = axis_stock_df[['Daily_Volatility']].sort_values(by='Daily_Volatility', ascending=False).head(5)

print("Top 5 periods of highest volatility (High - Low):")
print(top_5_volatility)

# Calculate rolling standard deviation of Close prices
axis_stock_df['Rolling_Std_Dev'] = axis_stock_df['Close'].rolling(window=30).std()

# Top 5 periods with highest rolling standard deviation
top_5_rolling_volatility = axis_stock_df[['Rolling_Std_Dev']].sort_values(by='Rolling_Std_Dev', ascending=False).head(5)

print("\nTop 5 periods of highest rolling volatility (Rolling Std Dev):")
print(top_5_rolling_volatility)
```

```python
import matplotlib.pyplot as plt

# Plot rolling standard deviation to visualize volatility
plt.figure(figsize=(14, 7))
plt.plot(axis_stock_df.index, axis_stock_df['Rolling_Std_Dev'], label='30-Day Rolling Std Dev', color='orange')
plt.scatter(top_5_rolling_volatility.index, top_5_rolling_volatility['Rolling_Std_Dev'], color='red', label='Top 5 High Volatility Periods')
plt.title('Stock Volatility Over Time')
plt.xlabel('Date')
plt.ylabel('Rolling Std Dev')
plt.legend()
plt.show()
```
